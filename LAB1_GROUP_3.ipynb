```python
# 8) Write the following function in the next cell.
def dummy(var):
    x = str(var)  # Convert the input variable to a string and assign it to 'x'
    y = len(x) - 1  # Calculate the index of the last character in 'x' and assign it to 'y'
    z = ''  # Initialize an empty string variable 'z'
    
    # Iterate over each character in 'x' in reverse order
    while y >= 0:
        z = z + x[y]  # Add the current character to the 'z' string
        y -= 1  # Move to the previous character
    
    return int(z)  # Convert the string 'z' back to an integer and return it

#Example with input 987654321:
print(dummy(987654321))
#The function converts the input 987654321 to a string and assigns it to x.
#y is initialized as the index of the last character in x, which is 8.
#An empty string z is created.
#The loop iterates in reverse order, appending each character of x to z in reverse.
#After the loop, z becomes '123456789'.
#The function converts '123456789' back to an integer and returns 123456789.
```

```python
#10) Following is the code for Calculate_Discount function which you implement in the Python Assessment.
def Calculate_Discount(price):
    if price >= 500:
        final_price = price * (70/100)
    elif price >= 350 and price < 500:
        final_price = price * (80/100)
    elif price >= 150 and price < 350:
        final_price = price * (90/100)
    elif price >= 10 and price < 150:
        final_price = price * (95/100)
    elif price > 0 and price < 10:
        final_price = price * (50/100)
    else:
        final_price = "Price cannot be negative. Please enter a positive value."
    return final_price

# Calculate the discount
# We note that this code is incorrect when price == 0.
# We would need to carefully check how this code is being used before changing the case for 0.
prices = [440, 265, 144, -10, 8]
for price in prices:
    discount = Calculate_Discount(price)
    if isinstance(discount, str):
        print(f"Price: {price:.2f}. {discount}")
    else:
        print(f"Price: {price:.2f}, Discounted Price: {discount:.2f}")
```

```python
#12) Implement the shift cipher in python
def caesar_encrypt(plaintext, shift):
    ciphertext = ""
    if plaintext.isalpha():  #validate alphabetic input
        for char in plaintext:
            if char.isupper():
                start = 65
            else:
                start = 97
            encrypted_char = chr((ord(char) - start + shift) % 26 + start)

            ciphertext += encrypted_char
    else:
        print("Only alphabetic input is allowed")

    return ciphertext


def caesar_decrypt(text, shift):
    # Shift is negative to reverse the encryption
    return caesar_encrypt(text, -shift)


plaintext = "cookie"
shift = 3

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)
print("----------")
plaintext = "USA"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "ISSEM"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)
print("----------")
plaintext = "CHICAGO"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "CAPPUCCINO"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "ZUPPA"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "XAVIERS"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "YELLOW"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "SECURITY"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "CYBERHUB"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)

print("----------")
plaintext = "HURRICANE"

encrypted_text = caesar_encrypt(plaintext, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)
```

```python
#13) What is hash function in cryptography?
## Cryptographic Hash Functions (CHFs) are mathematical functions that take an input of variable length, 
## apply a hash algorithm and return an output of fixed length
## CHFs can be used to verify digital signatures, message authentication codes (MACs), and other forms of authentication. 
## They are also commonly used to index data in hash tables, for fingerprinting binary files or data, to detect 
## duplicate data, and as checksums to detect accidental data corruption.
```

```python
#14) Write a hash function in Python to create the hash value of below message:
#Message = “Information Systems Security Engineering and Management”
import hashlib
def hash_message(message):
    #Raise an error if type is not string. We should not try to recover if different input types are given.
    #Trying to recover may create false expectations about what we can handle with this function, e.g. custom
    #objects that contain cyclic references and do not have __str__ implemented.
    if not isinstance(message, str):
        raise Exception("Input message must be of type string")
    
    #Actual string encoding shouldn't matter as long as two different strings don't result in the same bytes
    mbytes = bytes(message, 'utf-8')
    return hashlib.sha256(mbytes).hexdigest()

message = "Information Systems Security Engineering and Management"
original_message_hash = hash_message(Message)
print(original_message_hash)
```
